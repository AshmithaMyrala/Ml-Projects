```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
#data collection and analysis
#loading the dataset to pandas dataflow
diabetes_dataset=pd.read_csv('/content/diabetes.csv')
```

```python
#printing the first 5 rows of dataset
diabetes_dataset.head()
```

```python
#numbe rof rows and columns
diabetes_dataset.shape
```

```python
#getting the statistical measures of data
diabetes_dataset.describe()
```

```python
diabetes_dataset['Outcome'].value_counts()
```

```python
diabetes_dataset.groupby('Outcome').mean()
```

```python
# Drop rows where Outcome is NaN
diabetes_dataset = diabetes_dataset.dropna(subset=['Outcome'])
#separating the data and lables
X=diabetes_dataset.drop(columns='Outcome', axis=1)
Y=diabetes_dataset['Outcome']
```

```python
print(X)
print(Y)
```

```python
#data standardization
scalar=StandardScaler()
```

```python
scalar.fit(X)
```

```python
stadardized_data=scalar.transform(X)
```

```python
print(stadardized_data)
```

```python
X=stadardized_data
Y=diabetes_dataset['Outcome']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#traning the model
classifier=svm.SVC(kernel='linear')
```

```python
#training the support vector machine classifier
classifier.fit(X_train, Y_train)
```

```python
#Model Evaluation
#accuracy score in training data
X_train_prediction=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction, Y_train)

```

```python
print('Accuracy score of the training data:', training_data_accuracy)
```

```python
#accuracy score in test data
X_test_prediction=classifier.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of the test data:', test_data_accuracy)
```

```python
#making predictive system
input_data=(3,1,85,66,29,0,26.6,0.351,31)
#changing the input data as numpy array
input_data_as_numpy_array=np.asarray(input_data)
#reshape the array as we are predicting for one instance
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)

#standadaize input data
std_data=scalar.transform(input_data_reshaped)
print(std_data)

prediction=classifier.predict(std_data)
print(prediction)
if(prediction[0]==0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
